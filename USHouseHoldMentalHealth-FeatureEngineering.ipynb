```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import regex as re
```

```python
df=pd.read_csv("USHouseHoldMentalHealth-EDA.csv")
df.head(5)
```

```python
df.drop(['Unnamed: 0','index','Time Period Label','Confidence Interval','Suppression Flag'],axis=1,inplace=True)
df
```

```python
df.iloc[0]
```

### Converting the Object datatye to date time and sepating year, month and day 
    

```python
df['Time Period Start Date'] = pd.to_datetime(df['Time Period Start Date'])
df['Time Period End Date'] = pd.to_datetime(df['Time Period End Date'])
df.info()
```

```python
def fetch_day(data):
    return data.day
def fetch_month(data):
    return data.month
def fetch_year(data):
    return data.year
```

```python
df['TPS_Day']=df['Time Period Start Date'].apply(fetch_day)
df['TPS_Month']=df['Time Period Start Date'].apply(fetch_month)
df['TPS_Year']=df['Time Period Start Date'].apply(fetch_year)
df['TPE_Day']=df['Time Period End Date'].apply(fetch_day)
df['TPE_Month']=df['Time Period End Date'].apply(fetch_month)
df['TPE_Year']=df['Time Period End Date'].apply(fetch_year)
df
```

### Drop Time Period Start Date and Time Period End Date columns

```python
df.drop(['Time Period Start Date','Time Period End Date'],axis=1,inplace=True)
df
```

### Quartile Range must be splitted into 2 different columns

```python
def QRange(data,Qrange):
    x=re.findall("[0-9]+[.][0-9]+",data)
    return x[Qrange] 
    
```

```python
df['LowerQR'] = df['Quartile Range'].apply(lambda x: QRange(x,0))
df['HigherQR'] = df['Quartile Range'].apply(lambda x: QRange(x,1))
df
```

#### Drop Quartile Range column

```python
df.drop(['Quartile Range'],axis=1,inplace=True)
df
```

### Cleaning the phase column

```python
def PhaseCol(data):
    phase=re.findall("[0-9]",data)
    return phase[0]
```

```python
df['Phase']=df['Phase'].apply(PhaseCol)
df
```

```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

```python
for i in categorical_features:
    print(f"\n{df[i].nunique()} different categories are present in {i} column \n\n", df[i].unique())
    
    print("*"*50)
```

```python
# Changing object Dtype to Numeric

df[["HigherQR", "LowerQR","Phase"]] = df[["HigherQR", "LowerQR","Phase"]].apply(pd.to_numeric)
df
```

```python
EncodedDf = pd.get_dummies(df, columns = ['Indicator', 'Group', 'State', 'Subgroup'],prefix=None)
EncodedDf.to_csv("EncodedDF1.csv")
```
